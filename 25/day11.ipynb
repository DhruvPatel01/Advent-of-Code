```python
from collections import defaultdict

def mapL(f, xs): return [f(x) for x in xs]
```

```python

with open("./inputs/day11.txt") as f:
    input = [s.replace(':', '').split() for s in f]
    graph = defaultdict(list)
    for u, *vs in input:
        graph[u] = vs
```

```python
def n_paths(start, end):   
    def do_dfs(graph, stack, cache, u):
        if u in stack:
            raise ValueError("Cycle detected")
        if u == end:
            return 1
        if u not in cache:
            stack.append(u)
            paths_to_out = sum(do_dfs(graph, stack, cache, v) for v in graph[u])
            cache[u] = paths_to_out
            stack.pop()
        return cache[u]
    return do_dfs(graph, [], {}, start)
```
